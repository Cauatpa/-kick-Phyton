```python
casosInterior = 0
casosRGM = 0

for i in range(0,len(dfnan['Casos'])):
  if(dfnan['Grande região'][i] == 'Interior'):
    casosInterior = casosInterior + dfnan['Casos'][i]
  else:
    casosRGM = casosRGM + dfnan['Casos'][i]

total = casosInterior+casosRGM
```

```python
casos10 = RGM['Casos'].to_numpy()
cidades10 = RGM['Município'].to_numpy()
casos10in = cidadesInterior['Casos'].to_numpy()
cidades10in = cidadesInterior['Município'].to_numpy()
```

```python
plt.barh(cidades10in, casos10in, color = 'blue')
plt.xlabel("Quantidade de casos")
plt.ylabel("Municípios do interior")
plt.title("10 municípios com maior quantidade de casos de covid no interior")
plt.savefig('CasosInterior.png', dpi=200)
plt.show()
```
